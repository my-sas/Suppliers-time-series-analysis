```python
preprocessor = Preprocessor()
zpp4 = preprocessor.zpp4_preprocessing()
spec = preprocessor.spec_preprocessing()

zpp4.to_csv('../data/processed_data/zpp4.csv', index=False)
spec.to_csv('../data/processed_data/specs.csv', index=False)

# zpp4 = pd.read_csv('../data/processed_data/zpp4.csv')
# spec = pd.read_csv('../data/processed_data/specs.csv')

# pd.to_datetime(temp['spec_date'], format='%Y-%m-%d')
```

```python
# inner join zpp4 и spec, таким образом получаем, только те поставки,
# для которых есть данные о спецификации

data = pd.merge(zpp4, spec, how='inner', on=['id', 'id'], suffixes = ('', '_DROP'))
data = data.drop(['supplier_DROP', 'basis_DROP', 'spec_date_DROP'], 1)
data.head(2)
```

```python
# Распределение поставок по шкале, где 0 - начало спецификации,
# а 1 - конец поставок

data['lateness_percentage'] = list(map(
    lambda x: (x[1]-x[0])/(x[2]-x[0]),
    data[['spec_date', 'date', 'delivery_period_end']].values
))

# Удаление артефактов

# спецификация и конец поставки в тот же день, поэтому получился inf
data = data.drop(data.loc[data['lateness_percentage'] == np.inf].index)
# поставка раньше спецификации
data.loc[data['lateness_percentage'] < 0, 'lateness_percentage'] = 0.

plt.figure(figsize=(12, 6))
g = sns.histplot(data['lateness_percentage']).set(
    title='Распределение поставок по доле времени, оставшегося до завершения поставки',
    xticks = range(11)
)
```

```python
# Доля уже доставленного товара от всего законтрактованого договору

data['weight_percentage'] = data.groupby('id')['quantity'].cumsum() / data['volume_contracted']

# Удаление артефактов
# удалить volume_contracted равное 0
data = data.drop(data.loc[data['weight_percentage'] == np.inf].index)
```

```python
# можно посмотреть динамику поставки товара,
# 1 - товар доставлен в полном объёме

history = data[data['id'] == data['id'].unique()[133]]
x = history['lateness_percentage'].values
y = history['weight_percentage'].values
annotation = history['price_change'].values

# plot settings
plt.figure(figsize=(9, 6))
plt.xlabel('Время поставки')
plt.ylabel('Доля доставленного сырья')
plt.title('График поставок по спецификации')

plt.plot(x, y)

# scatter plot
d = max(abs(annotation.min()), annotation.max())
plt.scatter(x, y, c=annotation, cmap='bwr', vmin=-d, vmax=d)

# add cbar
cbar = plt.colorbar()
cbar.set_label('Качество сырья')
```

```python
supplier = data['supplier'].unique()[133]
history = data[data['supplier'] == supplier]

# проходимся по поставкам
x_list = []
y_list = []
for spec_id in history['id'].unique():
    
    spec_df = data.loc[data['id'] == spec_id]
    x = spec_df['lateness_percentage'].values
    y = spec_df['weight_percentage'].values
    
    x_list.append(x)
    y_list.append(y)

# получаем значения x
xnew = np.unique(np.concatenate(x_list))
xnew.sort(0)
```

```python
# продолжительность поставок в днях
data['delivery_period_lenght'] = (data['delivery_period_end'] - data['spec_date']).dt.days
sns.histplot(data['delivery_period_lenght'])
```

```python
# cmap = plt.cm.get_cmap('coolwarm')
# norm = mcolors.Normalize(vmin=0, vmax=4000)
# plt.plot(
#     x, y,
#     color=cmap(norm(history['volume_contracted'].iloc[0])),
#     alpha=0.06, linewidth=2
#     )

def multiplicity_plot(col, thresh):
    label_condition = lambda x: int(x[col].iloc[0] > thresh)
    labels = [f'{col} {sign} {thresh}' for sign in ('<', '>')] # < if 0, > if 1

    # plot settings
    cmap = plt.cm.get_cmap('bwr', 2)
    plt.title('Множественный график')
    plt.xlabel('Время поставки')
    plt.ylabel('Доля доставленного сырья')
    plt.xlim([0, 2])
    plt.ylim([0, 1.2])

    for supply_id in data['id'].unique():

        # collect supply history data
        history = data.loc[data['id'] == supply_id]
        label_num = label_condition(history)
        x = history['lateness_percentage'].values
        y = history['weight_percentage'].values

        plt.plot(
            x, y,
            color=cmap(label_condition(history)),
            alpha=0.03, linewidth=2
        )

    patches = [mpatches.Patch(color=cmap(i), label=label) for i, label in enumerate(labels)]
    plt.legend(handles=patches)
```

```python
multiplicity_plot('delivery_period_lenght', 25)
```

```python
multiplicity_plot('volume_contracted', 200)
```

```python
def multiplicity_plot1(col):
    # plot settings
    cmap = plt.cm.get_cmap('bwr', 2)
    plt.title('Множественный график')
    plt.xlabel('Время поставки')
    plt.ylabel('Доля доставленного сырья')
    plt.xlim([0, 2])
    plt.ylim([0, 1.2])
    
    for supply_id in data['id'].unique():
        
        # collect supply history data
        history = data.loc[data['id'] == supply_id]
        label_num = label_condition(history)
        x = history['lateness_percentage'].values
        y = history['weight_percentage'].values
        c = history[col].values

        plt.scatter(
            x, y, c=c,
            cmap='bwr', linewidths = 0.1,
            alpha=0.33, vmin=-10, vmax=10
        )
```
