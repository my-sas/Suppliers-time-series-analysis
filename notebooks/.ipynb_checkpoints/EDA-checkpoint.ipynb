```python
import random
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.patches as mpatches

from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder

import sys
import os
sys.path.append(os.path.abspath('../data'))
from preprocessing import Preprocessor

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
```

```python
preprocessor = Preprocessor()
zpp4 = preprocessor.zpp4_preprocessing()
spec = preprocessor.spec_preprocessing()

zpp4.to_csv('../data/processed_data/zpp4.csv', index=False)
spec.to_csv('../data/processed_data/specs.csv', index=False)

# zpp4 = pd.read_csv('../data/processed_data/zpp4.csv')
# spec = pd.read_csv('../data/processed_data/specs.csv')

# pd.to_datetime(temp['spec_date'], format='%Y-%m-%d')
```

```python
# inner join zpp4 и spec, таким образом получаем, только те поставки,
# для которых есть данные о спецификации

data = pd.merge(zpp4, spec, how='inner', on=['id', 'id'], suffixes = ('', '_DROP'))
data = data.drop(['supplier_DROP', 'basis_DROP', 'spec_date_DROP'], 1)
data.head(2)
```

```python
suppliers_history = data.groupby('supplier').apply(lambda x: x.sort_values('date'))
suppliers_history = pd.concat(
    (suppliers_history['date'],
     suppliers_history['quantity'].cumsum()),
    axis = 1
)
```

```python
supplier = data['supplier'].unique()[303]
for supplier in data['supplier'].unique()[0:1]:
    supplier_history = suppliers_history.loc[supplier]
    plt.plot(supplier_history['date'], supplier_history['quantity'], marker='o')
```

```python
"""
На сколько поставщик в процентном соотношении близок к концу спецификации
"""

data['lateness_percentage'] = list(map(
    lambda x: (x[1]-x[0])/(x[2]-x[0]),
    data[['spec_date', 'date', 'delivery_period_end']].values
))

# Удаление артефактов

# спецификация и конец поставки в тот же день, поэтому получился inf
data = data.drop(data.loc[data['lateness_percentage'] == np.inf].index)
# поставка раньше спецификации
data.loc[data['lateness_percentage'] < 0, 'lateness_percentage'] = 0.

plt.figure(figsize=(12, 6))
g = sns.histplot(data['lateness_percentage']).set(
    title='Распределение поставок по доле времени, оставшегося до завершения поставки',
    xticks = range(11)
)
```

```python
"""
Доля уже доставленного товара от всего законтрактованого договору
"""

data['weight_percentage'] = data.groupby('id')['quantity'].cumsum() / data['volume_contracted']

# Удаление артефактов
# удалить volume_contracted равное 0
data = data.drop(data.loc[data['weight_percentage'] == np.inf].index)
```

```python
# можно посмотреть динамику поставки товара,
# 1 - товар доставлен в полном объёме

history = data[data['id'] == data['id'].unique()[133]]
x = history['lateness_percentage'].values
y = history['weight_percentage'].values
annotation = history['price_change'].values

# plot settings
plt.figure(figsize=(9, 6))
plt.xlabel('Время поставки')
plt.ylabel('Доля доставленного сырья')
plt.title('График поставок по спецификации')

plt.plot(x, y)

# scatter plot
d = max(abs(annotation.min()), annotation.max())
plt.scatter(x, y, c=annotation, cmap='bwr', vmin=-d, vmax=d)

# add cbar
cbar = plt.colorbar()
cbar.set_label('Качество сырья')

plt.savefig('../figures/supplier_spec_history.jpg', bbox_inches='tight')
```

```python
"""
Временные фичи
"""

data['weekday'] = data['date'].map(lambda x: x.weekday())
data['month'] = data['date'].map(lambda x: x.month)
data['day'] = data['date'].map(lambda x: x.day)

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
g = sns.boxplot(data, y='price_change', x='month', showfliers=False).set(
    xticklabels=months,
    title = 'Разпределение доли изменения цены в зависимости от месяца'
)
plt.show()

g = sns.boxplot(data, y='price_change', x='day', showfliers=False).set(
    # xticklabels=months,
    title = 'Разпределение доли изменения цены в зависимости от дня недели'
)
plt.show()
```

```python
# продолжительность спецификации днях
data['delivery_period_lenght'] = (data['delivery_period_end'] - data['spec_date']).dt.days
sns.histplot(data['delivery_period_lenght'])
```

```python
def multiplicity_plot(col, thresh):
    label_condition = lambda x: int(x[col].iloc[0] > thresh)
    labels = [f'{col} {sign} {thresh}' for sign in ('<', '>')] # < if 0, > if 1

    # plot settings
    cmap = plt.cm.get_cmap('bwr', 2)
    plt.title('Множественный график')
    plt.xlabel('Время поставки')
    plt.ylabel('Доля доставленного сырья')
    plt.xlim([0, 2])
    plt.ylim([0, 1.2])

    for supply_id in data['id'].unique():

        # collect supply history data
        history = data.loc[data['id'] == supply_id]
        label_num = label_condition(history)
        x = history['lateness_percentage'].values
        y = history['weight_percentage'].values

        plt.plot(
            x, y,
            color=cmap(label_condition(history)),
            alpha=0.03, linewidth=2
        )

    patches = [mpatches.Patch(color=cmap(i), label=label) for i, label in enumerate(labels)]
    plt.legend(handles=patches)
```

```python
def multiplicity_plot_labels(col):
    labels = data[col].unique()
    label_idx = {k: v for k, v in zip(labels, range(len(labels)))}
    cmap = plt.cm.get_cmap('bwr', len(labels))
    # colors = [cmap(label_idx[i]) for i in y]
    
    
    plt.title('Множественный график')
    plt.xlabel('Время поставки')
    plt.ylabel('Доля доставленного сырья')
    plt.xlim([0, 2])
    plt.ylim([0, 1.2])
    
    for supply_id in data['id'].unique():

        # collect supply history data
        history = data.loc[data['id'] == supply_id]
        x = history['lateness_percentage'].values
        y = history['weight_percentage'].values
        c = cmap(label_idx[history[col].iloc[0]])

        plt.plot(
            x, y, color=c,
            alpha=0.03, linewidth=2
        )

    patches = [mpatches.Patch(color=cmap(i), label=label) for i, label in enumerate(labels)]
    plt.legend(handles=patches)
```

```python
"""
Последняя дата отгрузки
"""

data['last_date'] = data.groupby(['id'])['date'].transform('max')
```

```python
"""
Сколько у поставщика одновременных спецификаций
"""

spec_times = data.groupby(['supplier', 'id'])['date'].agg(['min', 'max']).reset_index()
data['supplier_active_specs'] = data.apply(
    lambda row: ((spec_times['supplier'] == row['supplier']) & 
                 (spec_times['min'] <= row['date']) & 
                 (row['date'] <= spec_times['max'])
                ).sum(), axis=1
)

spec_times = data.groupby(['id'])['date'].agg(['min', 'max']).reset_index()
data['active_specs'] = data.apply(
    lambda row: ((spec_times['min'] <= row['date']) & 
                 (row['date'] <= spec_times['max'])).sum(), axis=1
)
```

```python
def set_last(group):
    group.iloc[-1] -= 1
    return group

data['supplier_active_specs'] = data.groupby('id')['supplier_active_specs'].apply(set_last)
data['active_specs'] = data.groupby('id')['active_specs'].apply(set_last)
```

```python
"""
Отношение пройденного времени (lateness_percentage) к доставленному грузу (weight_percentage)
Близость к 0 означает, что поставщик доставил много груза за единицу времени.
Высокое значение указывает на то, что поставщик не успевает.
Единица означает линейность.
"""

data['performance'] = data['lateness_percentage'] / data['weight_percentage']
```

```python
# Колонки, которые будут использоваться
id_cols = cols = ['supplier', 'id']
learn_cols = [
    'supplier_status', 'volume_contracted', 'spec_price', 'delivery_period_lenght', 'active_specs', 'last_date',# параметры спецификации
    'price_change', 'lateness_percentage', 'weight_percentage', # параметры поставок
    'month', 'day', 'weekday'
]

learn_df = data[id_cols + learn_cols]

learn_df = learn_df.drop(learn_df[learn_df.isnull().any(axis=1)].index)

scaler = StandardScaler()
learn_df[['volume_contracted', 'spec_price', 'price_change', 'delivery_period_lenght', 'price_change']] = scaler.fit_transform(learn_df[['volume_contracted', 'spec_price', 'price_change', 'delivery_period_lenght', 'price_change']])
learn_df['price_change'] = learn_df['price_change']*0.4


learn_df['supplier_status'] = (learn_df['supplier_status'].values == 'СХТП').astype(int)

learn_df.to_csv('../data/final_data/data.csv', index=False)
```
